```python
import matplotlib.pyplot as plt

# データ
k = list(range(1, 11))
cv_error = [3.5061, 4.0224, 3.8718, 4.0654, 3.9578, 4.0224, 4.0009, 4.2160, 4.3235, 4.2160]
testdata_error = [2.7748, 3.3556, 2.7318, 3.0974, 3.2695, 3.5276, 3.6782, 3.6567, 3.7643, 3.7427]

# グラフを描画
plt.figure(figsize=(8, 6))
plt.plot(k, cv_error, marker='x', linestyle='-', color='blue', label='cv')
plt.plot(k, testdata_error, marker='x', linestyle='--', color='red', label='testdata')

# グラフのラベルとタイトル
plt.xlabel('$k$')
plt.ylabel('misclass rate [%]')
plt.legend()

# y軸の範囲
plt.ylim(2.5, 4.5)

# グリッド線
plt.grid(True)

# グラフを保存
plt.savefig('fig_2.pdf')

# グラフを表示
plt.show()
```
